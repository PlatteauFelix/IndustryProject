```python
from PIL import Image
from skimage import data, io, filters
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
from sklearn import manifold, datasets

dataDir=Path('C:\\Users\\viers\\OneDrive\\Bureaublad\\MCT\\sem4\\IndustryProject\\IndustryProject\\goals-2\\train')
annFile = Path('C:\\Users\\viers\\OneDrive\\Bureaublad\\MCT\\sem4\\IndustryProject\\IndustryProject\\goals-2\\train\\_annotations.coco.json')

def toTif(dirName):
    dataDir = f'C:\\Users\\viers\\OneDrive\\Bureaublad\\MCT\\sem4\\IndustryProject\\IndustryProject\\goals-2\\{dirName}'
    annFile = f"{dataDir}\\_annotations.coco.json"
    coco = COCO(annFile)
    imgIds = coco.getImgIds()
    imgs = coco.loadImgs(imgIds)
    pics = []
    
    for img in imgs:
        pics.append(img['file_name'])

    for i in range(0,len(imgIds)):
        path=pics[i]
        img= io.imread(f"{dataDir}\\"+path)

        path=str(i)+'.tif'
        io.imsave(f"{dataDir}\\"+path,img)
        
    for i in range(0,len(imgIds)):
        path=pics[i]
        img= io.imread(f"{dataDir}\\masks\\"+path)

        path=str(i)+'.tif'
        io.imsave(f"{dataDir}\\masks\\"+path,img)

toTif("train")
```
